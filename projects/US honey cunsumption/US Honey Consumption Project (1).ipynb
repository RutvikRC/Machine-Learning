## Top Honey Producing States 

```python
df2 = df.sort_values("production", ascending= False)
df2.head(20)
```

```python
df2
```

```python
#States having the Highest no.of Stocks
```

```python
plt.figure(figsize=(15,10))
sns.barplot(y=df2["state"],x=df2["stocks"])
plt.ylabel(" Year(1995-2021)")
plt.xlabel(" No.of Stocks")
plt.show()
```

```python
# Inference- From the above figure we can infer that Top 3 states having large no. of stocks are:
#     North Dakoda
#     South Dakoda
#     California
```

## States having the Highest no.of Colonies
