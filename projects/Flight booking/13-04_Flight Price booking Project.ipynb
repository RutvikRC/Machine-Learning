```python
#  8) Convert all the attributes into integer data type which are of object data type except
#Flight column.

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

df["airline"] = le.fit_transform(df["airline"])
df["source_city"] = le.fit_transform(df["source_city"])
df["departure_time"] = le.fit_transform(df["departure_time"])
df["arrival_time"]=le.fit_transform(df["arrival_time"])
df["destination_city"]=le.fit_transform(df["destination_city"])
df["class"]=le.fit_transform(df["class"])
df['stops'] = le.fit_transform(df['stops'])
```

```python
df.info()
```

```python
df = df.drop(columns=['flight'])
```

```python
#Mutlicollinearity - vif

from statsmodels.stats.outliers_influence import variance_inflation_factor 
col_list = []
for col in df.columns:
    if ((df[col].dtype != 'object') & (col != 'price') ):
        col_list.append(col)
        
col_list
```

```python
X = df[col_list]
vif_data = pd.DataFrame()
vif_data['feature'] = X.columns
vif_data["VIF"] = [variance_inflation_factor(X.values, i) for i in range(len(X.columns))] 
print(vif_data)
```

```python
#Decision Tree
X = df.drop(columns=['price'])
y = df['price']
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=78)
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
```

```python
x_train = sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import *

dt=DecisionTreeRegressor()
dt.fit(x_train,y_train) #training
y_pred=dt.predict(x_test) #predicting values
r2_score(y_test,y_pred)
```

```python
sns.distplot(y_test,label="Actual")
sns.distplot(y_pred,label="Predicted")
plt.title("decision Tree")
plt.legend()
```

```python
#MSE
#RMSE
```

```python
#Random Forest
from sklearn.ensemble import RandomForestRegressor
rfr=RandomForestRegressor()
rfr.fit(x_train,y_train) #training
y_pred=rfr.predict(x_test) #predicting
r2_score(y_test,y_pred)
```
